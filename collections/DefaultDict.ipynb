```python
from collections import defaultdict

# Never raise key error
```

```python
dd = defaultdict(list)
dd[1] = 'teja'
dd['ambati'] = 'teja'
dd['u'] = 99
dd['list'] = [1,2,4]
print(dd['xyz'])
```

```python
dd = defaultdict(set)
dd[1] = 'teja'
dd['ambati'] = 'teja'
dd['u'] = 99
dd['list'] = [1,2,4]
print(dd['xyz'])
```

```python
dd = defaultdict(int)
```

```python
dd[1] = 'teja'
dd['ambati'] = 'teja'
dd['u'] = 99
dd['list'] = [1,2,4]
```

```python
print(dd['list'])
```

```python
print(dd['xyz'])
```

```python
dd[[1,5,6]] = 'xyz'
```

```python
dd = defaultdict(list)
dd["teja"] = [7,8,9]
dd['teja'].append(10)
print(dd['teja'])
```
