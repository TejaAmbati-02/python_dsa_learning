```python
c1 = Counter([ 1, 3, 2, 2, 2, 3, 3, 4])
c2 = Counter([ 3, 3, 1, 4, 5])
```

```python
c3 = c1 - c2 # Remove the elements of c2 form c1
```

```python
list(c3.elements())
```

```python
c3 = c2 - c1

print(list(c3.elements()))
```

```python
c3 = c1 + c2
print(list(c3.elements()))
# Follows insertion order of c1
```

```python
c3 = c2 + c1
print(list(c3.elements()))
# Follows insertion order of c2
```

```python
c3 = c2 & c1
print(list(c3.elements()))
# Common between 2 is taken
```

```python
c3 = c1 & c2
print(list(c3.elements()))
# Common between 2 is taken
# Follow insertion order of c1
```

```python
c3 = c2 & c1
print(list(c3.elements()))
# Common between 2 is taken
# Follow insertion order of c2
```

```python
c3 = c2 | c1
print(list(c3.elements()))
# Everything between 2 is taken
# Follow insertion order of c2
```
