```python
from collections import deque
```

```python
dq = deque([2,3,1])
dq.append(5)
print(dq)
```

```python
dq.appendleft(7)
print(dq)
```

```python
print(dq.pop())
print(dq)
```

```python
print(dq.popleft())
print(dq)
```

```python
dq.extend([10, 'sai'])
print(dq)
```

```python
dq.extendleft([89, 'teja'])
print(dq)
```

```python
dq.rotate(1)
print(dq)
```
